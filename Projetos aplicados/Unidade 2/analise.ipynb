```python
import pandas as pd
dados = pd.read_csv('base_projeto02.csv')
```

```python
dados
```

```python
resultados = pd.DataFrame(columns = ['nome_modelo','f1_score','accuracy_score'])
```

```python
resultados
```

```python
dados = dados.dropna()
```

```python
dados = dados[dados['gender'].isin(['Male','Female'])]
dados
```

```python
dados = dados[dados.vehicle_age.isin(['1-2 Year','< 1 Year','> 2 Years'])]
dados
```

```python
dados = pd.get_dummies(dados,columns=['gender','vehicle_age'],dtype=int)
dados
```

```python
import numpy as np
dados['vehicle_damage'] = np.where(dados['vehicle_damage']=='Yes',1,0)
```

```python
dados
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dados_escalonados = pd.DataFrame(scaler.fit_transform(dados),columns=dados.columns)
```

```python
dados_escalonados
```

```python
colunas = dados.columns.tolist()
colunas.remove('response')
colunas.append('response')
dados_escalonados = dados_escalonados[colunas]
dados_escalonados.info()
```

```python
treino = dados_escalonados.sample(frac = 0.8)
teste = dados_escalonados.drop(index=treino.index)
```

```python
treino.shape,teste.shape,dados_escalonados.shape
```

```python
x_treino = treino.iloc[:,:13].values
y_treino = treino.iloc[:,13].values
x_teste = teste.iloc[:,:13].values
y_teste = teste.iloc[:,13].values
```

```python
x_treino
```

```python
from sklearn.metrics import f1_score,accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
modelo = DecisionTreeClassifier()
modelo.fit(x_treino,y_treino)
y_pred = modelo.predict(x_teste)
f1 = f1_score(y_teste,y_pred)
acc = accuracy_score(y_teste,y_pred)
print(f"Esse modelo retornou F1 Score = {f1} e Accuracy Score = {acc}")
```

```python
nome = input('Digite o nome do modelo: ')
add = [nome,f1,acc]
resultados.loc[resultados.shape[0]]=add
resultados
```
