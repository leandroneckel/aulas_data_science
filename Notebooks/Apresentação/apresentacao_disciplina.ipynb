```python
total_linhas = dados.shape[0]
for coluna in dados.columns:
    total_nulos_coluna = dados[coluna].isnull().sum()
    pc_nulos_coluna = round(total_nulos_coluna/total_linhas*100,2)
    print(f'A coluna "{coluna}" apresenta {total_nulos_coluna} registros nulos, que representam {pc_nulos_coluna} % do total de registros')

```

```python
total_linhas = dados.shape[0]
coluna_nome = list()
total_nulo = list()
pc_total_nulo = list()
for coluna in dados.columns:
    coluna_nome.append(coluna)
    total_nulos_coluna = dados[coluna].isnull().sum()
    total_nulo.append(total_nulos_coluna)
    pc_nulos_coluna = round(total_nulos_coluna/total_linhas*100,2)
    pc_total_nulo.append(pc_nulos_coluna)
```

```python
resumo = pd.DataFrame()
resumo['Nome da coluna'] = coluna_nome
resumo['Total nulos'] = total_nulo
resumo['Total nulos (pc)'] = pc_total_nulo
```

```python
resumo
```

```python
resumo = resumo.sort_values(by='Total nulos (pc)')
```

```python
resumo
```

```python
resumo = resumo.sort_values(by='Total nulos (pc)',ascending=False)
resumo
```
