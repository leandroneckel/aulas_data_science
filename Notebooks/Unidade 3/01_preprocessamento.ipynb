# Machine Learning Classificação
## Abertura e tratamento de dados
Obs.: Originalmente, o dataframe contém classificação da qualidade do vinho com notas discretas

```python
import pandas as pd
dados = pd.read_excel('dados_vinho_original.xlsx')
dados.info()
```

```python
dados.quality.value_counts()
```

Vamos transformar interpretar o problema como um problema de classifição de somente duas classes: Vinho bom ou vinho ruim, atribuindo 0 para notas entre 3 e 6 e 1 para notas maiores ou iguais a 7

```python
import numpy as np
dados['quality'] = np.where(dados['quality']<7,0,1)
dados.quality.value_counts()
```

```python
dados
```

Executando restante do preprocessamento:
* Fazendo dummy encoding na coluna type

```python
dados = pd.get_dummies(dados,columns=['type'])
dados
```

* Fazendo o escalonamento dos dados por meio da normalização

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
dados_escalonados = pd.DataFrame(scaler.fit_transform(dados),columns=dados.columns)
dados_escalonados
```

* Organizando as colunas

```python
colunas = dados_escalonados.columns.tolist()
colunas.remove('quality')
colunas.append('quality')
dados_escalonados = dados_escalonados[colunas]
dados_escalonados.info()
```

Salvando base de dados para prosseguir com o Machine Learning

```python
dados_escalonados.to_excel('dados_vinho_tratado.xlsx',index=False)
```
