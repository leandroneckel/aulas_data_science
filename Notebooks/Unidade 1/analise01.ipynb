## Colunas
1. age
2. sex
3. chest pain type (4 values)
4. resting blood pressure
5. serum cholestoral in mg/dl
6. fasting blood sugar > 120 mg/dl
7. resting electrocardiographic results (values 0,1,2)
8. maximum heart rate achieved
9. exercise induced angina
10. oldpeak = ST depression induced by exercise relative to rest
11. the slope of the peak exercise ST segment
12. number of major vessels (0-3) colored by flourosopy
13. thal: 1 = normal; 2 = fixed defect; 3 = reversable defect (0 - no registry)

```python
#calcular desvios e desvios quadráticos da idade
idade = dados[['age']]
idade_media = idade.age.mean() #calcula a média
idade['desvios_absolutos'] = abs(idade.age -idade_media)
```

```python
idade['desvios_quadraticos'] = idade.desvios_absolutos**2
```

## Analisando a variável Thal (categórica)

```python
dados.thal
```

```python
plt.hist(dados.thal)
```

```python
#Cálculo da moda
dados.thal.mode()
```

```python
#valor esperado: para calcular, precisamos transformar para valores não nulos
thal = dados[['thal']]
#filtrando para valores não nulos
thal = thal[thal.thal!=0]
plt.hist(thal.thal)
```

```python
#calculando as probabilidades de cada categoria e organizando em uma tabela
probs = thal.thal.value_counts(normalize=True)
df_probs = probs.to_frame()
df_probs.reset_index(inplace=True)
df_probs.columns = ['categoria', 'probabilidade']
df_probs
```

```python
df_probs['momento'] = df_probs.categoria*df_probs.probabilidade
df_probs
```

```python
valor_esperado = df_probs.momento.sum()
valor_esperado
```

```python
# Gráfico de dispersão de idade contra trestbps ressaltanto com cores o chol
x = dados.age
y = dados.trestbps
cor = dados.chol
sc = plt.scatter(x, y, c=cor, cmap='viridis') 
plt.colorbar(sc, label='chol')
plt.xlabel('age')
plt.ylabel('trestpbs')
plt.grid(ls = ':')
```

```python
#colocar um quarto elemento no gráfico como o tamanho sendo a variável oldpeak
min_val = dados.oldpeak[dados.oldpeak != 0].min()
dados.oldpeak = dados.oldpeak.replace(0, min_val * 0.1)

tamanho = dados.oldpeak
plt.figure(figsize=(10, 10))  
sc = plt.scatter(x, y, c=cor, s=tamanho*100, cmap='viridis', alpha=0.6) 
plt.colorbar(sc, label='chol')
plt.xlabel('age')
plt.ylabel('trestpbs')
plt.grid(ls = ':')

sizes_legend = [round(np.min(tamanho),1), np.max(tamanho)]
labels = [str(s) for s in sizes_legend]

legend_elements = [plt.Line2D([0], [0], marker='o', color='w', label=label,
                              markerfacecolor='gray', markersize=s*5) 
                   for label, s in zip(labels, sizes_legend)]
plt.legend(handles=legend_elements, title="oldpeak")
```

```python
#grafico de contorno
from scipy.interpolate import griddata
xi = np.linspace(dados['age'].min(), dados['age'].max(), 100)
yi = np.linspace(dados['trestbps'].min(), dados['trestbps'].max(), 100)
xi, yi = np.meshgrid(xi, yi)
zi = griddata((dados['age'], dados['trestbps']), dados['chol'], (xi, yi), method='cubic')
plt.figure(figsize=(10, 10)) 
plt.contour(xi, yi, zi, levels=14, linewidths=0.5, colors='k')
plt.contourf(xi, yi, zi, levels=14, cmap=plt.cm.rainbow)
plt.colorbar(label='chol')
plt.xlabel('age')
plt.ylabel('trestpbs')
```
