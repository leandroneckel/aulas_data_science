# Padronização ou Normalização de dados

Muitos algoritmos de Machine Learning não lidam bem com dados em faixas diferentes. Assim, se torna necessário fazer a padronização ou a normalização dos dados

## Padronização (Standard)

Leva em conta a média e o desvio padrão. Mais recomendado para dados distribuídos sob uma normal

## Normalização (MinMax)

Leva em conta o valor máximo e o valor mínimo

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
# Padronizando
scaler = StandardScaler()
dados_padrao = pd.DataFrame(scaler.fit_transform(dados), columns=dados.columns)
dados_padrao
````

````python
# Normalização
min_max_scaler = MinMaxScaler()
dados_normal = pd.DataFrame(min_max_scaler.fit_transform(dados), columns=dados.columns)
dados_normal
````

# Detecção de outliers

Outliers podem afetar significantemente o desempenho de um modelo de Machine Learning. Assim, fazer a detecção e tratamento já na faze de pré-processamento é importante. 

Entrentanto, é importante ter em mente que o tratamento de outliers pode afetar muito significantemente a distribuição dos dados e afastá-los do seu estado original (algo que tentamos evitar ao tratar valores faltantes)

Assim, é importante que sejam feitos testes sobre tratar os outliers antes ou depois de fazer a separação de dados em dados de treino e dados de teste. O mais recomentado é após a separação dos dados em dados de treino e dados de teste.

Façamos a separação manual primeiramente então

## Separação em base de treino e base de teste

````python
treino = dados_normal.sample(frac = 0.80)
teste = dados_normal.drop(treino.index)
````

## Tratamento de outliers

### Opção 1 - Coluna a coluna

Somente para colunas com larga distribuição de dados (colunas booleanas, categoricas gerais ou ordinais não entram) por meio da inputação de valores no lugar dos outliers (risco de alterar distribuição original)

````python
treino.describe()
````

No dataframe a única coluna elencável é a ```age```

Um método de tratar os outliers é inputar valores novos para todos os dados que não estejam dentro da distância inter quartil ou estabelecer um percentil específico para o corte

````python
import matplotlib.pyplot as plt
plt.boxplot(treino.age);
````

````python
q1 = treino['age'].quantile(0.25)
q3 = treino['age'].quantile(0.75)
print(f'Quartil inferior: {q1}. Quartil superior: {q3}')
````

````python
p_inf = treino['age'].quantile(0.05)
p_sup = treino['age'].quantile(0.95)
print(f'Percentil inferior: {p_inf}. Percentil superior: {p_sup}')
````

Filtrando e estimando a média e desvio da população para inputar aleatoriamente dados segundo a distribuição

````python
inliers = treino[(treino['age'] >= p_inf ) & (treino['age'] <= p_sup )]
media = inliers.age.mean()
std = inliers.age.std()
import numpy as np
def generate_random_age():
    random_age = np.random.normal(media,std)
    return int(random_age)
def alterar_idade(linha):
    if (linha['age'] < p_inf ) | (linha['age'] > p_sup ):
        return generate_random_age()
    else:
        return linha['age']
treino_copy = treino.copy()
````

````python
treino_copy['age'] = treino_copy.apply(alterar_idade,axis=1)
````

````python
plt.boxplot(treino_copy.age);
````

````python
treino_copy.age.describe()
````

## Opçao 2 - Eliminação de outliers por meio de z-scores

Nesse caso é necessário definir qual é o limite do z-score aceito para filtragem

````python
from scipy import stats
import numpy as np

z_scores = stats.zscore(treino)
abs_z_scores = np.abs(z_scores)
outliers = (abs_z_scores > 3).all(axis=1)
treino[outliers]
````

# Opção 3 - Eliminação de outliers por meio de machine learning - Isolation Forest

Neste caso é necessário informar qual é o percentual da base de dados que se estima estar contaminado

````python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(contamination=0.05)
preds = clf.fit_predict(treino)
treino_result = treino.copy()
treino_result['inlier'] = preds
outliers = treino_result[treino_result.inlier == -1]
outliers
````

# Opção 4 - Eliminação de outliers por meio de machine learning - DBScan

Nesse caso, é necessário informar a distância mínima entre duas amostras para serem consideradas vizinhas (eps) e o número de amostras em uma vizinhança para um ponto ser considerado como um ponto central (min_samples). É um algoritmo baseado em distâncias

````python
treino
````

````python
from sklearn.cluster import DBSCAN

outlier_detection = DBSCAN(min_samples = 2, eps = 3)
clusters = outlier_detection.fit_predict(treino)
treino_result = treino.copy()
treino_result['inlier'] = clusters
outliers = treino_result[treino_result.inlier == -1]
outliers
````
