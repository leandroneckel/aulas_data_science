```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dados = pd.read_csv('ideia_06/car.data',header=None)
colunas = ['buying','maint','doors','persons','lug_boot','safety','evalutation']
dados.columns = colunas
df = dados.copy()
```

```python
c0 = dados.buying.unique().tolist()
c0.sort()
c0
```

```python
c0d = [2,0,1,3]
dfc0 = pd.DataFrame()
dfc0['names'] = c0
dfc0['numbers'] = c0d
dc = pd.Series(dfc0.numbers.values, index=dfc0.names).to_dict()
dc
```

```python
df['buying'] = df['buying'].map(dc)
df
```
